```python
import docx
import pycld2 as cld2
import glob
```

```python
list_names = []
path = './*.docx'
file_names = glob.glob(path)

for file in file_names:
    file = file.replace(path.replace('*.docx', ''), '')    
    list_names.append(file)
    # print(file)
```

```python
for name in list_names:
    print('\n---------------------------------------------')
    print('Analyzing: ', name)
    print('---------------------------------------------')
    doc = docx.Document(name)
    count = 0
    for i in doc.paragraphs:
        lines = i.text.split('.')
        for line in lines:
            count += 1
            reliable, index, top_3_choices,vecs = cld2.detect(line, returnVectors=True)
            if top_3_choices[0][0] != 'ENGLISH':
                # print(line)
                # print(top_3_choices[0][0])
                if top_3_choices[0][0] != 'Unknown':
                    print(line)
                    print(top_3_choices[0][0],' -> NON-ENGLISH CONTENT DETECTED!!!!!')
                    # break;
    if count < 10:
        print("\n", count, "THE DOCUEMNT IS TOO SHORT -> POSSIBLE NON-PP FOUND!!!!!")
    print('---------------------------------------------')
    print('Finished')
    print('---------------------------------------------')

print("Done")
```
